```python
import pandas as pd
from pathlib import Path
import numpy as np
import seaborn as sns
```

```python
nflx_data = Path("/Users/macjohnson/Documents/UMN_Fintech/classwork/python_analysis/nflx.csv")
```

```python
nflx_df = pd.read_csv(nflx_data, index_col="date", infer_datetime_format=True, parse_dates=True)
nflx_df.sort_index(inplace=True)
nflx_df.head()
```

```python
#daily closing prices plot 
nflx_df.plot()
```

```python
#Rolling 30 day mean nflx closing prices
nflx_df.rolling(window=30).mean().plot()
```

```python
#Plot a rolling 180-day std dev of TSLA's closing prices
nflx_df.rolling(window=180).std().plot()
```

```python
# daily closing prices of Netflix
ax = nflx_df.plot()

#180-Day Rolling Mean on the same figure
nflx_df.rolling(window=180).mean().plot(ax=ax)

# Set the legend of the figure
ax.legend(["NFLX", "NFLX 180 Day Mean"]);
```
